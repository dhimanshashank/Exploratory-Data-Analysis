# Marvel Exploratory Data Analysis

```python
#Loading the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
marvel = pd.read_csv("marvel_characters_stats.csv")
```

```python
marvel.head()
```

```python
marvel["Alignment"].value_counts()
```

## Trying to Fin

```python
good_marvel = marvel[marvel['Alignment'] == 'good']
```

```python
good_marvel.head()
```

```python
good_marvel.sort_values(by = 'Speed', ascending = False).head()
```

```python
max_speed_good = good_marvel[good_marvel['Speed'] == 100]
```

```python
max_speed_good
```

```python
min_speed_good = good_marvel[good_marvel['Speed'] == 1]
```

```python
min_speed_good
```

```python
good_marvel.sort_values(by = 'Power', ascending = False).head()
```

```python
good_max_power_full = good_marvel[good_marvel['Power'] == 100]
```

```python
good_max_power_full.head()
```

```python
good_max_power_full.shape
```

```python
good_max_power = good_marvel.sort_values(by = ['Total'], ascending = False)
```

```python
good_max_power
```

```python
plt.figure(figsize=(8,5))
plt.bar(list(good_max_power['Name'])[0:5], list(good_max_power['Total'])[0:5], color = 'g')
plt.show()
```

```python
#Making a histogram for speed for good superheros
plt.figure(figsize = (10,5))
plt.hist(good_marvel['Speed'])
plt.title("Distribution of speed")
plt.xlabel('Speed')
plt.show()
```
